```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import cm as cmx
from matplotlib import colors
import matplotlib as mpl
```

```python
class SimpleRNG:
    def __init__(self, m, a, c, seed):
        self.seed = seed
        self.m = m
        self.a = a
        self.c = c
        
    def draw(self):
        self.seed = (self.a * self.seed + self.c) % self.m
        return self.seed
```

```python
def rngseq(rng, N):
    seq = []
    for i in range(N):
        seq.append(rng.draw())
    return seq
```

```python
def plot_sequence_2D(seq):
    x = seq[::2]
    y = seq[1::2]
    size = min(len(x), len(y))
    x = x[:size]
    y = y[:size]
    plt.scatter(x,y,s=5)
    plt.show()
```

```python
rng = SimpleRNG(1377, 13, 7, 3)
seq = rngseq(rng,1400)
plot_sequence_2D(seq)
```

```python
def rngseq2(rng1, rng2, N):
    seq = []
    for i in range(N):
        val1 = rng1.draw()
        val2 = rng2.draw()
        while not (val2%2==0 or val2%3==0 or val2%5==0 or val2%7==0 or val2%13==0):
            val1 = rng1.draw()
            val2 = rng2.draw()
        seq.append(val1)
    return seq
```

```python
rng1 = SimpleRNG(5127,249,7,7)
rng2 = SimpleRNG(5127,249,7,7)

seq1 = rngseq(rng,4000)
plot_sequence_2D(seq1)

seq = rngseq2(rng1, rng2, 4000)
plot_sequence_2D(seq)
```

```python
def integrate(func, borders, N, algorithm = 'trapezoidal'):
    x = np.linspace(borders[0], borders[1], N)
    h = ((borders[1]-borders[0]) / (N - 1))
    if algorithm == 'trapezoidal':
        weights = np.ones_like(x)
        weights[[0,-1]] = 0.5
    if algorithm == 'simpson':
        weights = np.ones_like(x)
        weights[1::2] = 4.0/3
        weights[2::2] = 2.0/3
        weights[[0,-1]] = 1.0/3
    result = np.sum(func(x) * weights*h)
    return result
    
```

```python
(integrate(np.sin, (0, np.pi), 10), integrate(np.sin, (0, np.pi), 10, 'simpson'))
```
