```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def logistic_map(x0, mu, iterations=200, dims=1, last=False):
    seq = [x0]
    for _ in range(iterations):
        x0 = mu * x0 * (1 - x0)
        seq.append(x0)
    return seq[-1] if last else seq
```

```python
x_0 = 0.9

u_list = np.linspace(0.5, 4, 8)

for u in u_list:
    seq = logistic_map(x_0, u, 30)
    plt.plot(seq, label=str(u))
    plt.legend()
    plt.show()
```

```python
logistic_iters = 200
mu_iters = 10000
mu_inter = (1, 4)
mu_list = np.random.uniform(mu_inter[0], mu_inter[1], mu_iters)
x0_list = np.random.random((mu_iters))
print(mu_list.shape)
print(x0_list.shape)

logistic_seq = logistic_map(x0_list, mu_list, logistic_iters, last=True)

plt.scatter(mu_list, logistic_seq, s=0.05)
```

```python
# dY/dt = F, t = time, h = time-step
def RK4(Y, F, t, h):
    k1 = h * F(t, Y)
    k2 = h * F(t + h / 2,Y + k1 / 2)
    k3 = h * F(t + h / 2,Y + k2 / 2)
    k4 = h * F(t + h, Y + k3)
    return Y + k1 / 6 + k2 / 3 + k3 / 3 + k4 / 6

# conersion to system of first order ODEs
def Fpend(t, Y):
    F = np.zeros_like(Y)
    F[0] = Y[1]
    F[1] = -(omega0 ** 2) * np.sin(Y[0]) - alpha * Y[1] + f * np.cos(omega * t)
    return F
```

```python
# free pendulum, no friction, no external force, small displacements
omega0 = 1
omega = 5
f = 0
alpha = 0
h = 0.1

orbits = []

for _ in range(100):
    print(f"calculating orbit {_}",end='\r')

    points = []
    Y = (np.random.random((2))-0.5)*2
    points.append(Y)
    t= 0.0
    tmax = 30
    while t < tmax:
        t += h
        Y = RK4(Y,Fpend,t,h)
        points.append(Y)
    points = np.asarray(points).transpose()
    
    orbits.append(points)
    
fig = plt.figure(dpi=120)
for points in orbits:
    plt.plot(points[0], points[1])
# plt.xlim(-10, 10)
plt.gca().set_aspect('equal')
```

```python
# conersion to system of first order ODEs
# a vibrating pivot
def Fpend_vib(t, Y):
    F = np.zeros((2))
    F[0] = Y[1]
    F[1] = -(omega0 ** 2 + f2 * np.cos(omega2 * t)) * np.sin(Y[0]) - alpha * Y[1] + f * np.cos(omega * t)
    return F
```

```python
# free pendulum, with friction, dominationg force, large displacements and velicoties
# causes chaos in some cases
omega0 = 1
omega = 0.5
f = 0
alpha = 0.5
h = 0.1
f2 = 10
omega2 = 0.5

orbits = []

for _ in range(100):
    print(f"calculating orbit {_}",end='\r')

    points = []
    Y = 10 * (np.random.random((2))-0.5)*2
    points.append(Y)
    t= 0.0
    tmax = 30
    while t < tmax:
        t += h
        Y = RK4(Y,Fpend_vib,t,h)
        points.append(Y)
    points = np.asarray(points).transpose()
    
    orbits.append(points)
    
fig = plt.figure(dpi=120)
for points in orbits:
    plt.plot(points[0], points[1])
# plt.xlim(-10, 10)
# plt.gca().set_aspect('equal')
```

```python
# bifurcation
# slow version

omega0 = 1
omega = 0.5
f = 0
alpha = 0.5
h = 0.1
omega2 = 0.5

f2param = np.linspace(0, 2, 100)

finpoints = []
for f2 in f2param:
    print(f"calculating orbit {f2}",end='\r')

    Y = np.pi*(np.random.random((2))-0.5)*2
    Y[0] = 0.0
    finpoints.append(Y)
    t = 0.0
    tmax = 30
    while t < tmax:
        t += h
        Y = RK4(Y,Fpend_vib,t,h)
        if Y[0] <= 0.1:
            finpoints[-1] = Y
    
finpoints = np.asarray(finpoints).transpose()
    
plt.scatter(f2param,np.abs(finpoints[1]),s=0.1)
# plt.xlim(-10, 10)
# plt.gca().set_aspect('equal')
```

```python
# enabling numpy to speed up by vector operations
def Fpend_vib_improved(t, Y):
    F = np.zeros_like(Y)
    F[0] = Y[1]
    F[1] = -(omega0 ** 2 + f2 * np.cos(omega2 * t)) * np.sin(Y[0]) - alpha * Y[1] + f * np.cos(omega * t)
    return F
```

```python
omega0 = 1
omega = 2
f = 0
alpha = 0.1
h = 0.1
omega2 = 0.5

f2 = np.linspace(0, 2, 100000)
vels = np.zeros_like(f2)
Y = 2 * (np.random.random((2, f2.size)) - 0.5)
# constant velocities for all orbits to make it a bit more deterministic
Y[1][:] = 1.0
t = 0.0
tmax = 100

while t < tmax:
    t += h
    Y = RK4(Y, Fpend_vib_improved, t, h)
    # getting velocities when angular displacement is 0
    cond = np.where(Y[0] <= 0.1)
    vels[cond] = Y[1][cond]
    

plt.scatter(f2, np.abs(vels), s=0.001)
```
