```python
import numpy as np

def energy(conf):
    return np.sum(np.multiply(conf[:-1], conf[1:]))
    
def metropolis(ising, n, k_T):
    config = [ising]
    for k in range(n):
        rand = np.random.randint(0, 50)
        trial = config[k].copy()
        trial[rand] = -trial[rand]
        if np.random.uniform(0,1) <= (np.exp((energy(config[k]) - energy(trial)) / k_T)):
            config.append(trial)
        else:
            config.append(config[k])
    return config
```

```python
ising = np.random.randint(0,2,50)
ising[ising == 0] = -1

k_T = 0.1
n = 500

config_history = metropolis(ising, n, k_T)
energies = [energy(config) for config in config_history]
print(energies)
```
